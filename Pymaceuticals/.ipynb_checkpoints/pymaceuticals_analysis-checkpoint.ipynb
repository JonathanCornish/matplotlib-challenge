```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load 
mouse_drug_data_to_csv = "data/mouse_drug_data.csv"
clinical_trial_data_to_csv = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv("data/mouse_drug_data.csv")
mouse_df.head()
clinical_df = pd.read_csv("data/clinicaltrial_data.csv")
clinical_df.head()

# Combine the data into a single dataset
combined_df = pd.merge(mouse_df, clinical_df, on="Mouse ID", how="outer")

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df = combined_df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_vol_df = pd.DataFrame(grouped_df)

# Flatten the column headings
tumor_vol_df = tumor_vol_df.reset_index()

# Preview DataFrame
tumor_vol_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
vol_sem = pd.DataFrame(combined_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"])
vol_sem.head()
```

```python
# Flatten the column headings
vol_sem.reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_vol_df = tumor_vol_df.reset_index()
tumor_vol_mean = tumor_vol_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

vol_sem = vol_sem.reset_index()
tumor_vol_sem = vol_sem.pivot(index='Timepoint', columns="Drug")['Tumor Volume (mm3)']

# Preview that Reformatting worked
tumor_vol_mean.head()
```

```python
# show the column headers for dataframe 'vol_sem'
list(vol_sem)
```

```python
# show the column headers & datatypes for the dataframe 'vol_sem'
vol_sem.columns
```

```python
# show the column headers for dataframe 'dm_df_sem'
list(tumor_vol_sem)
```

```python
# show the column headers for dataframe 'dm_df'
list(tumor_vol_mean)
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_vol_mean.index, tumor_vol_mean['Capomulin'], yerr=tumor_vol_sem['Capomulin'], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vol_mean.index, tumor_vol_mean["Infubinol"], yerr=tumor_vol_sem["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vol_mean.index, tumor_vol_mean["Ketapril"], yerr=tumor_vol_sem["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vol_mean.index, tumor_vol_mean["Placebo"], yerr=tumor_vol_sem["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("Tumor Volume Change (%)")
plt.xlabel("Treatment Duration (Days)")
plt.xticks(rotation = 45)
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Figures/Fig1.png")
plt.show()
```

```python
# Flatten the column headings & un-groupby, essentially
m_sem = m_sem.reset_index()

# Display the top 15 results
m_sem.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames
m_sem_pivot = m_sem.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
m_mean_pivot = m_df_mean.pivot(index="Timepoint", columns = "Drug")["Metastatic Sites"]

# Preview that Reformatting worked
m_mean_pivot
```

```python
# Preview that Reformatting worked
m_sem_pivot
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(m_mean_pivot.index, m_mean_pivot['Capomulin'], yerr=m_sem_pivot['Capomulin'], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(m_mean_pivot.index, m_mean_pivot["Infubinol"], yerr=m_sem_pivot["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(m_mean_pivot.index, m_mean_pivot["Ketapril"], yerr=m_sem_pivot["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(m_mean_pivot.index, m_mean_pivot["Placebo"], yerr=m_sem_pivot["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Metastatic Spread Over 45 Day Treatment")
plt.ylabel("Metastatic Sites")
plt.xlabel("Treatment Duration (Days)")
plt.xticks(rotation = 45)
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Figures/Fig2.png")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_count = combined_df.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame & rename the column 'Tumor Volume (mm3)' to 'Number_of_Mice'
survival_count = pd.DataFrame({"Number_of_Mice": survival_count})

# Preview DataFrame
survival_count.head().reset_index()
```

```python
survival_count
```

```python
# data-munging to re-format the data frame
survival_count = survival_count.reset_index()
survival_count_pivot = survival_count.pivot(index="Timepoint", columns="Drug")["Number_of_Mice"]

# preview the dataframe
survival_count_pivot
```

```python
list(survival_count_pivot)
```

```python
list(survival_count)
```

```python
# show the number of mice remaining over time for each drug i.e. show the number of surviving mice at each time period
cap_surv_list = survival_count_pivot.loc[:,'Capomulin'].tolist()
inf_surv_list = survival_count_pivot.loc[:,'Infubinol'].tolist()
ket_surv_list = survival_count_pivot.loc[:,'Ketapril'].tolist()
plc_surv_list = survival_count_pivot.loc[:,'Placebo'].tolist()


# List comprehension to get the mouse survival rate
# Each value in the lists is divided by the first value of the respective list to get a percentage
cap_survrate = [(x/cap_surv_list[0])*100 for x in cap_surv_list]
inf_survrate = [(x/inf_surv_list[0])*100 for x in inf_surv_list]
ket_survrate = [(x/ket_surv_list[0])*100 for x in ket_surv_list]
plc_survrate = [(x/plc_surv_list[0])*100 for x in plc_surv_list]

cap_survrate
```

```python
cap_surv_list
```

```python
ax = plt.subplot(111)

# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(0, 50, 5)

# Set the plot title and axes titles
plt.title("Survival During Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Survival Rate (%)")

# Plot the 'mean' list vs. the established x axis with error 
ax.plot(x_axis, cap_survrate, color="g", label="Capomulin")
ax.plot(x_axis, inf_survrate, color="r", label="Infubinol")
ax.plot(x_axis, ket_survrate, color="y", label="Ketapril")
ax.plot(x_axis, plc_survrate, color="blue", label="Placebo")

# Add the legend and gridlines
ax.legend(loc=3)

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.xlim(0, max(x_axis))

# Save the Figure
plt.savefig("Figures/Fig3.png")

# Show the resulting scatter plot
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_perc_change = 100 * (tumor_vol_mean.iloc[-1] - tumor_vol_mean.iloc[0]) / tumor_vol_mean.iloc[0]
tumor_perc_change_sem = 100 * (tumor_vol_sem.iloc[-1] - tumor_vol_sem.iloc[0]) / tumor_vol_sem.iloc[0]

# Display the data to confirm
tumor_perc_change
```

```python
tumor_perc_change_sem
```

```python
# print(f"{tumor_perc_change} {tumor_perc_change_sem}")
# print(f"{tumor_perc_change_sem}")
# print(f"{tumor_perc_change}")
tumor_vol_df.head(15)
drug_unique = tumor_vol_df["Drug"].unique()
drug_unique
i=9
y=0
tup=[]
for drug in drug_unique:
    tumor_per_change = 100 * (tumor_vol_df.iloc[i,2] -  tumor_vol_df.iloc[y,2])/ tumor_vol_df.iloc[y,2]
    print(f"{drug} {tumor_per_change}")
    tup.append({drug})
    tup.append({tumor_per_change})
    i+=10
    y+=10
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_changes = (tumor_perc_change["Capomulin"], 
               tumor_perc_change["Infubinol"], 
               tumor_perc_change["Ketapril"], 
               tumor_perc_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_changes))  
width = 1
rectsPass = ax.bar(ind[0], pct_changes[0], width, color='green')
rectsFail = ax.bar(ind[1:], pct_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Save the Figure
fig.savefig("Figures/Fig4.png")

# Show the Figure
fig.show()
```

```python
# Concluding thoughts/summary:
# Overall the trends indicate that Capomulin and Ramicane are the only drugs that showed efficacy in reducing 
# the volume of tumors found in the tested mice.

# Capomulin and Ramicane also produced the highest survival rates in the days following the start of treatment. 

# there is reason to believe that lower tumor volumes is correlated with greater life expectancy. Since
# Ramicane and Capomulin appear to reduce the tumor volume found in mice, there is reason to believe this is the
# reason why survival rates are also higher for these drugs.

# However, none of the drugs had a large enough sample size of mice for these results to be statistically 
# significant, given that n<40 for every single drug in terms of the number of mice observed. This impairs
# our ability to draw firm conclusions from the test data. 
```
